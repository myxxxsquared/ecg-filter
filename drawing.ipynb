```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_origin = np.array(list(map(float, filter(None, map(str.strip, open('ecg.txt'))))))
data_processed = np.array(list(map(float, filter(None, map(str.strip, open('ecg_processed.txt'))))))
```

```python
assert data_origin.shape == data_processed.shape
data_origin.shape
```

```python
xs = np.arange(data_origin.shape[0])
plt.figure(figsize=(10, 10))
print("origin")
plt.plot(xs, data_origin)
plt.show()

plt.figure(figsize=(10, 10))
print("processed")
plt.plot(xs, data_processed)
# plt.axis((3000, 6000, -0.5e7, 0.5e7))
plt.show()
```
